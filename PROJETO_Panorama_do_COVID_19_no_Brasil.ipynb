# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada pelo recém-descoberto coronavírus. E vem causando estragos, tanto econômico quanto também social, matando muitas pessoas, aumentando o desemprego, e consequentemente a fome ao redor no mundo.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

<center><img alt="Colaboratory logo" width="50%" src="https://raw.githubusercontent.com/carlosfab/dsnp2/master/img/covid19.jpg"></center>

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela, porém esse grupo pode continuar transmitindo o vírus, o que acarreta em um aumento no número de casos.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

<center><img width="50%" src="https://image.freepik.com/vetores-gratis/plano-de-fundo-de-vacina-de-coronavirus-desenhado-a-mao-plana_23-2148868964.jpg"></center>

Com o avanço da tecnologia e da ciência, algumas vacinas foram criadas, e aparentemente os casos da doença, onde a vacinação está avançando, estão diminuindo e vem aumento a esperança para a volta da normalidade.


Porém nesse meio tempo, surgiram novas variantes, mais transmissíveis , e que por falta de dados não sabemos se são mais letais ou se as vacinas atuais funcionam contra elas, e podem colocar em cheque todo o avanço das vacinações.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados
Os Dados foram retirados deste [link](https://github.com/owid/covid-19-data/tree/master/public/data), e são da Our World in Data, que é uma publicação digital especializada em dados analíticos sobre mudanças nas condições de qualidade de vida ao redor do mundo.


A Our World in Data colhem diariamente os dados de fontes oficias dos governos, publicações científicas, dados de pesquisa de universidades e fazem uma curadoria e com isso criaram esse dataset que iremos analisar, portanto é uma base de dados bem completa.

Por se tratar de uma doença nova e de uma situação praticamente inédito na história recente, as lacunas de informações e conhecimento ainda são muito grandes, informações como: taxas de letalidade, potencial de transmissão, tratamento, existencia de outros efeitos ou sequelas no organismo dos que foram infectados. Todas essas informaçõs ainda são preliminares, pois o tempo de estudo é muito pequeno.

Nesse momento, a produção científica é crucial para melhor compreender a doença e seus efeitos e buscar soluções. Pesquisadores e cientistas, no mundo todo, em muitos casos a partir de uma boa coordenação governamental, estão se mobilizando para estimar tanto os efeitos da doença sobre a saúde da população quanto os impactos econômicos e sociais dessa pandemia. A Organização Mundial de Saúde (OMS), por exemplo, tem coordenado e mapeado os esforços de pesquisa no mundo, reunindo cientistas no tema e identificado as prioridades de pesquisa neste momento.


### Dicionário de Variáveis

O dataset possui 63 Variáveis:

* iso_code: Codigo em padrão internacional para identificação dos países
* continent: Continente onde está localizado
* location: localização geográfica
* date: data da observação
* total_cases: total de casos confirmados
* new_cases: novos casos 
* new_cases_smoothed: novos casos (média móvel de 7 dias)
* total_deaths: Total de mortes atribuidas ao covid
* new_deaths: Novas mortes atribuidas ao Covid
* new_deaths_smoothed:	Novas mortes atribuidas ao covid (média móvel de 7 dias)
* total_cases_per_million: Número total de casos por 1 milhão de pessoas
* new_cases_per_million: Novos casos por 1 milhão de pessoas
* new_cases_smoothed_per_million: Novos casos por 1 milhão de pessoas (média móvel de 7 dias)
* total_deaths_per_million: Total de mortes por 1 milhão de pessoas
* new_deaths_per_million: Novas mortes por 1 milhão de pessoas
* new_deaths_smoothed_per_million: Novas mortes por 1 milhão de pessoas (média móvel de 7 dias)
* reproduction_rate: Estimativa de reprodução da COVID em tempo real
* icu_patients: Pacientes na UTI no dia
* icu_patients_per_million: Pacientes na UTI no dia por 1 milhão de pessoas
* hosp_patients: Pacientes no hospital por covid
* hosp_patients_per_million: Pacientes no hospital por 1 milhão de pessoas
* weekly_icu_admissions: Número de entradas semanais na UTI
* weekly_icu_admissions_per_million: Número de entradas semanais na UTI por milhão.
* weekly_hosp_admissions: Número de entradas no hospital pela Covid
* weekly_hosp_admissions_per_million: Número de entradas no hospital pela Covid por milhão

Podemos verificar que a coluna 'date'está como objeto, portanto irei transforma-lo no formato datetime:

Podemos observar uma grande porcentagem de dados faltantes, a princípio poderíamos pensar que os dados e sua fonte são ruins, porém a análise deve ir um pouco mais fundo que isso:
* Temos que levar em conta que alguns países nem começaram a vacinação, portanto o total de vacinados estará ausente;

* Países como o Brasil em que a testagem não faz parte do plano de governo, acaba influenciando nos números dos dados.

Como dito anteiormente, é uma situação nova e nem todos os países estão devidamente preparados para coletar dados por isso existem alguns valores ausentes.

```
# Irei criar uma copia do Dataset para poder manipular melhor os dados, pois caso haja um erro, é mais fácil de recuperar.
df1 = df.copy()
```

```
# Filtrarei os valor do Mês passado
df1 = df1.loc[df1.date == '2021-07-17']
```

```
# Coloquei como índice o iso_code para poder manipular melhor
df1.set_index('iso_code', inplace=True)
```

```
# Verificar atualmente (último mes) quais são os países com mais casos confirmados e mais mortes:

df1.sort_values(by=['total_cases'], ascending=False)
```

Existem dados de  continentes inteiros, mundo inteiro, e isso prejudica a visão de saber quais os paises mais infectados, por isso irei excluir esses dados para ficar apenas com os países.

```
df1.drop(['OWID_WRL', 'OWID_ASI', 'OWID_EUR', 'OWID_NAM', 'OWID_SAM', 'OWID_EUN', 'OWID_AFR'], inplace=True)
```

```
df1.sort_values(by=['total_cases'], ascending=False).head()
```

Por essa Análise, os países com mais casos são:
* Estados Unidos da América
* Índia
* Brasil
* França
* Russia

```
df1.sort_values(by=['total_deaths'], ascending=False).head()
```

Aqui conseguimos ver que o ranking mudou:
* Estados Unidos da América
* Brasil
* Índia
* Mexico
* Peru

Isso mostra que :
1. Não necessariamente que o país com mais infecções vai matar mais pessoas
2. Países com mais pessoas vacinadas, podem até contrair a doença, mas matam menos, provando a eficácia da vacina;
3. O numero de habitantes do país pode interferir nesse resultado, pois quanto maior o número, maior serão os dados.

Agora irei plotar os gráficos com os 5 países em que mais morreram pessoas em decorrencia da COVID no mes passado.



```
df_d = df1[['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[:5]

fig, ax = plt.subplots()

df_d.plot(x='location', y='total_deaths', kind='barh', label='total de mortos', figsize=(10,6), ax=ax);
ax.set_title('Total de mortos')
ax.set_ylabel('Países')
ax.set_xlabel('Número de mortos')
ax.legend()

plt.tight_layout()
```

Agora irei verificar a evolução de mortes no mundo.

```
fig, ax = plt.subplots()

df.loc[df.location == "World", "total_deaths",].reset_index(drop=True).plot(figsize=(10,6), ax=ax)
```

### Análise Exploratória para o Brasil


```
df_br = df.loc[df.location == 'Brazil'].copy()
```

```
fig, ax = plt.subplots()

df_br.plot(x='date', y='new_cases_smoothed', kind='line', ax=ax, label='Novos casos confirmados', figsize=(10,6))

ax.set_title('Novos casos confirmados de COVID-19 (Média de 7 dias)')
ax.legend()

plt.tight_layout()
```

O Gráfico mostra os novos casos diários confirmados de COVID-19, ela mostra em uma média móvel de 7 dias.
É importante notar que os dados do dia não necessariamente são as reais, pois há o atraso de envio de informação e repasse ao público.

Mas já podemos observar uma tendência de queda no número de novos casos.

```
fig, ax = plt.subplots()

df_br.plot(x='date', y='new_deaths_smoothed', kind='line', ax=ax, label='Mortes diárias por COVID-19', figsize=(10,6))

ax.set_title('Mortes diárias de COVID-19 (Média de 7 dias)')
ax.legend()

plt.tight_layout()
```

O gráfico vai mostrar a média móvel de 7dias do número de mortes diárias por covid-19.
As mortes mostradas em um dia, não necessariamente mostram o numero de mortes naquele dia, mas sim as mortes reportadas no dia.

Aqui percebemos uma queda Brusca nas mortes, isso devido provavelmente à vacina.


```
fig, ax = plt.subplots()

df_br.loc[df_br['date'] > '2020-12-01'].plot(x='date', y='new_vaccinations_smoothed', kind='line', ax=ax, label='Aplicações diárias de Vacina', figsize=(10,6))

ax.set_title('Aplicação diária de vacina (Média de 7 dias)')
ax.legend()

plt.tight_layout()
```

O gráfico reforça que a vacinação está andando a passos largos no país, podemos ver algumas coisas interessantes:
* A vacinação de fato, começou por volta de janeiro de 2021 em ritmo lento
* Como visto em noticiários, houve poucas doses, atraso na entrega, e isso se mostra no gráfico, onde possui alguns gaps entre vacinação de um dia para o outro
* Existe um período de estagnação, seguido de um aumento de doses aplicadas.
